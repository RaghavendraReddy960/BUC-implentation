```python
#Used it later in the algo
output = {}
min_sup = 500
    
```

```python
# 1 index array used it int the last 
df['last'] = df.apply(lambda x: 1, axis = 1)
```

```python
data = df.to_numpy()
```

```python
d1 = pd.DataFrame(data=data)
d1.info()
```

```python
col_no = d1.columns[-1]
```

```python
def BUC_Implementation(df, prefix=None):
    if len(df.columns) == 1: # len columns=1 returning
        return

    if prefix is None: #saving prefixes
        prefix = []

    for i in df.columns: # main Buc Loops
        if i == df.columns[-1]:
            continue
        count = df.groupby([i]).count() #count wtr to groupby
        count = count[count >= min_sup].dropna() #Droping with null attributes and checking > min attribute and adding it if passes

        for j in count.index:
            subdf = df[df[i].isin([j])]
            subdf = subdf.drop(i,1)

            BUC_Implementation(subdf,prefix+[str(j)]) #Recursively performing BUC on lower dimensions
            output[",".join(prefix+[str(j)])] = int(count[col_no][j]) #printing the final values.

        df = df.drop(i,1)
```

```python
result = BUC_Implementation(d1)
```

```python
for a,b in output.items():
    print(a,b)
```
